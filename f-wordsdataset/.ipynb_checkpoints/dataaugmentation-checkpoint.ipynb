```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

import IPython

file_name = "C://Users//SMU//Desktop//practice.wav"
data, sr = librosa.load(file_name)

IPython.display.Audio(data, rate=sr)
```

```python
# waveshow 그래프
plt.figure(figsize=(12, 4))
librosa.display.waveshow(data, sr=sr)
plt.axis("off")
plt.savefig("./waveshow_original.png", bbox_inches="tight", pad_inches=0)
plt.show()
```

```python
# noise 추가된 waveshow 그래프
noise_amp = np.amax(data) * 0.2 * np.random.uniform()
noise = noise_amp * np.random.normal(size=data.shape)

data_noise = data + noise
plt.figure(figsize=(12, 4))
librosa.display.waveshow(data_noise, sr=sr)
plt.axis("off")
plt.show()
```

```python
rate = 0.8 + np.random.random() * 0.4  # 최소 80%, 최대 120% 비율로 스트레칭
data_stretch = librosa.effects.time_stretch(data, rate=rate)
plt.figure(figsize=(12, 4))
librosa.display.waveshow(data_stretch, sr=sr)
plt.axis("off")
plt.show()
```

```python
# stft
stft = librosa.stft(data)
stft_db = librosa.amplitude_to_db(abs(stft), ref=np.max)

plt.figure(figsize=(12, 4))
librosa.display.specshow(stft_db, sr=sr, x_axis="time", y_axis="hz")
plt.show()
```

```python
noise_amp = np.amax(data) * 0.05 * np.random.uniform()
noise = noise_amp * np.random.normal(size=data.shape)
data_noise = data + noise

stft_noise = librosa.stft(data_noise)
stft_db_noise = librosa.amplitude_to_db(abs(stft_noise), ref=np.max)

plt.figure(figsize=(12, 4))
librosa.display.specshow(stft_db_noise, sr=sr, x_axis="time", y_axis="hz")
plt.show()
```

```python
rate = 0.8 + np.random.random() * 0.4  # 최소 80%, 최대 120% 비율로 스트레칭
data_stretched = librosa.effects.time_stretch(data, rate=rate)

stft_stretched = librosa.stft(data_stretched)
stft_db_stretched = librosa.amplitude_to_db(np.abs(stft_stretched), ref=np.max)
plt.figure(figsize=(12, 4))
librosa.display.specshow(stft_db_stretched, sr=sr, x_axis="time", y_axis="hz")
plt.show()
```

```python
mel_spec = librosa.feature.melspectrogram(S=np.abs(stft))
mel_spec_db = librosa.amplitude_to_db(mel_spec, ref=np.max)

plt.figure(figsize=(12, 4))
librosa.display.specshow(mel_spec_db, sr=sr, x_axis="time", y_axis="mel")
plt.show()
```

```python
mel_spec_noise = librosa.feature.melspectrogram(S=np.abs(stft_noise))
mel_spec_db_noise = librosa.amplitude_to_db(mel_spec_noise, ref=np.max)

plt.figure(figsize=(12, 4))
librosa.display.specshow(mel_spec_db_noise, sr=sr, x_axis="time", y_axis="mel")
plt.show()
```

```python
mel_spec_stretched = librosa.feature.melspectrogram(S=np.abs(stft_stretched))
mel_spec_db_stretched = librosa.amplitude_to_db(mel_spec_stretched, ref=np.max)

plt.figure(figsize=(12, 4))
librosa.display.specshow(mel_spec_db_stretched, sr=sr, x_axis="time", y_axis="mel")
plt.show()
```

```python
# 정사각형 사이즈로 리사이즈 (흑색 패딩 붙임)
import cv2

target_size = (255, 255)

def padding_to_square(img: np.ndarray, square_size: int) -> np.ndarray:
    h, w, c = img.shape

    if h == w:
        bg = img
    
    if h < w:
        padding_size = (w - h) // 2
        bg = np.zeros((w, w, c), dtype=np.uint8)
        bg[padding_size:padding_size + h, :] = img.copy()

    else: # h > w
        padding_size = (h - w) // 2
        bg = np.zeros((h, h, c), dtype=np.uint8)
        bg[:, padding_size : padding_size + w] = img.copy()
    
    result = cv2.resize(bg, (square_size, square_size))
    return result
    
image = cv2.imread("waveshow_original.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
padded_img = padding_to_square(image, 255)
plt.imshow(padded_img)
plt.show()
```
